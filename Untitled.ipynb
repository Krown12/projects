```python
df.head()
```

```python
plt.subplots(figsize=(10, 10)) 
for i in range(12):
    plt.subplot(4, 3, i+1)
    k = np.random.randint(0,len(df))
    img = cv2.imread(df.loc[k,'filepath'])
    plt.imshow(img)
    plt.title(df.loc[k,'breed'])
    plt.axis('off')
plt.show()    
```
